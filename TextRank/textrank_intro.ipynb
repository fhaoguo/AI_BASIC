### What is the PageRank value for node 0 (node with id 0) after one PageRank iteration?
### pagerank equation: $r_j = \sum_{i \rightarrow j} \beta \frac {r_i} {d_i} + (1 - \beta) \frac {1}{N}$

```python
# 计算每个节点的度数
node_degree = {}
# 计算每个节点的邻居
node_neighbors = {}
for node in g.nodes():
    node_degree[node] = g.degree(node)
    node_neighbors[node] = list(g.neighbors(node))

# 查看每个节点的度数信息
node_degree
```

```python
# 查看每个节点的邻居信息
node_neighbors
```

```python
# 某个节点的一次迭代更新方法
def one_iter_pagerank(g, beta, r0_kv, node_id):
    r1 = 0.0
    for n_nid in node_neighbors[node_id]:
        r1 += beta * r0_kv[n_nid] / node_degree[n_nid]
    r1 += (1 - beta) * 1 / g.number_of_nodes()
    
    return r1
```

```python
# 初始化参数值
beta = 0.8
r0 = 1 / g.number_of_nodes()
r_kv = {}
for nid in list(g.nodes()):
    r_kv[nid] = r0
```

```python
# 迭代n次，计算所有节点的pagerank值，并查看disp_id节点的每次迭代计算结果
n = 50
disp_id = 10
for r in range(n):
    r_kv_tmp = {}
    for nid in list(g.nodes()):
        r_kv_tmp[nid] = one_iter_pagerank(g, beta, r_kv, nid)
        if nid == disp_id:
            print("The initial PageRank value for node {} is {}".format(nid, r_kv[nid]))
            print("The PageRank value for node {} after one iteration is {}".format(nid, r_kv_tmp[nid]))
    
    r_kv = r_kv_tmp.copy()
    r_kv_tmp.clear()
```
