```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# tanh函数
def tanh(x):
    y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    return y
```

```python
# 绘制tanh图
def plot_tanh():
    # 设置参数x（起点，终点，间距）
    x = np.arange(-8, 8, 0.2)
    y = tanh(x)
    plt.plot(x, y)
    plt.grid()
    plt.show()
```

```python
print(tanh(4))
```

```python
plot_tanh()
```

```python
def derivate_tanh(x):
    
    return 1 - np.power(tanh(x), 2)
```

```python
def plot_derivate_tanh():
    # 设置参数x（起点，终点，间距）
    x = np.arange(-8, 8, 0.2)
    y = derivate_tanh(x)
    plt.plot(x, y)
    plt.grid()
    plt.show()
```

```python
plot_derivate_tanh()
```
