```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sigmoid函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# sigmoid函数的导数
def derivate_sigmoid(x):
    s = sigmoid(x)
    return s * (1 - s)
```

```python
# 绘制sigmoid图
def plot_sigmoid():
    # 设置参数x（起点， 终点， 间距）
    x = np.arange(-8, 8, 0.2)
    y = sigmoid(x)
    plt.plot(x, y)
    plt.grid()
    plt.show()
```

```python
# 绘制sigmoid函数的导数
def plot_ds():
    # 设置参数x（起点， 终点， 间距）
    x = np.arange(-8, 8, 0.2)
    y = derivate_sigmoid(x)
    plt.plot(x, y)
    plt.show()
```

```python
print(sigmoid(4))
```

```python
plot_sigmoid()
```

```python
plot_ds()
```
